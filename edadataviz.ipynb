### TASK  3: Visualize the relationship between success rate of each orbit type


Next, we want to visually check if there are any relationship between success rate and orbit type.


Let's create a `bar chart` for the sucess rate of each orbit


```python
# Group by the 'Orbit' column and calculate the mean success rate (Class)
orbit_success_df = df.groupby('Orbit')['Class'].mean().reset_index()

# Create a bar chart using Matplotlib
plt.figure(figsize=(10,6))
plt.bar(orbit_success_df['Orbit'], orbit_success_df['Class'], color='skyblue')

# Customize the axis labels and title
plt.xlabel('Orbit Type', fontsize=15)
plt.ylabel('Success Rate', fontsize=15)
plt.title('Success Rate by Orbit Type', fontsize=18)
plt.xticks(rotation=45)

# Display the plot
plt.show()
```

```python
import matplotlib.cm as cm
import numpy as np

# Group by the 'Orbit' column and calculate the mean success rate (Class)
orbit_success_df = df.groupby('Orbit')['Class'].mean().reset_index()

# Sort the orbit types by success rate in descending order
orbit_success_df = orbit_success_df.sort_values(by='Class', ascending=False)

# Add a ranking column based on the success rate, ensuring the rank starts at 1
orbit_success_df['Rank'] = orbit_success_df['Class'].rank(ascending=False, method='dense').astype(int)

# Create a colormap to color the bars based on rank
cmap = cm.get_cmap('Blues', np.max(orbit_success_df['Rank']) - np.min(orbit_success_df['Rank']) + 1)
colors = [cmap(rank - 1) for rank in orbit_success_df['Rank']]

# Create a bar chart using Matplotlib
plt.figure(figsize=(10,6))
plt.bar(orbit_success_df['Orbit'], orbit_success_df['Class'], color=colors)

# Customize the axis labels and title
plt.xlabel('Orbit Type', fontsize=15)
plt.ylabel('Success Rate', fontsize=15)
plt.title('Success Rate by Orbit Type (Ranked)', fontsize=18)
plt.xticks(rotation=45)

# Display the ranking information above each bar
for i, value in enumerate(orbit_success_df['Class']):
    plt.text(i, value + 0.01, f'Rank: {int(orbit_success_df["Rank"].iloc[i])}', ha='center', fontsize=12)

# Display the plot
plt.show()

# Display the ranked dataframe
print(orbit_success_df)
```

Analyze the plotted bar chart to identify which orbits have the highest success rates.


### TASK  4: Visualize the relationship between FlightNumber and Orbit type


For each orbit, we want to see if there is any relationship between FlightNumber and Orbit type.


```python
# Plot a scatter point chart with x axis to be FlightNumber and y axis to be the Orbit, and hue to be the class value
```

```python
# Import necessary libraries
import matplotlib.pyplot as plt

# Create the scatter plot
plt.figure(figsize=(12,6))

# Use scatter plot with FlightNumber on x-axis, Orbit on y-axis, and color by Class (success/failure)
scatter = plt.scatter(df['FlightNumber'], df['Orbit'], c=df['Class'], cmap='coolwarm', edgecolor='k', s=100)

# Customize axis labels and title
plt.xlabel('Flight Number', fontsize=15)
plt.ylabel('Orbit Type', fontsize=15)
plt.title('Relationship between Flight Number and Orbit Type', fontsize=18)

# Add color legend to distinguish between success (1) and failure (0)
plt.colorbar(scatter, label='Class (Success/Failure)')

# Display the plot
plt.show()
```

You can observe that in the LEO orbit, success seems to be related to the number of flights. Conversely, in the GTO orbit, there appears to be no relationship between flight number and success.


### TASK  5: Visualize the relationship between Payload Mass and Orbit type


Similarly, we can plot the Payload Mass vs. Orbit scatter point charts to reveal the relationship between Payload Mass and Orbit type


```python
# Plot a scatter point chart with x axis to be Payload Mass and y axis tobe the Orbit, and hue to be the class value
# Import necessary libraries
import matplotlib.pyplot as plt

# Create the scatter plot
plt.figure(figsize=(12,6))

# Use scatter plot with Payload Mass on x-axis, Orbit on y-axis, and color by Class (success/failure)
scatter = plt.scatter(df['PayloadMass'], df['Orbit'], c=df['Class'], cmap='coolwarm', edgecolor='k', s=100)

# Customize axis labels and title
plt.xlabel('Payload Mass (kg)', fontsize=15)
plt.ylabel('Orbit Type', fontsize=15)
plt.title('Relationship between Payload Mass and Orbit Type', fontsize=18)

# Add color legend to distinguish between success (1) and failure (0)
plt.colorbar(scatter, label='Class (Success/Failure)')

# Display the plot
plt.show()
```

```python
# Plot a line chart with x axis to be the extracted year and y axis to be the success rate
```

```python
# A function to extract years from the Date column
def Extract_year():
    year = []
    for i in df["Date"]:
        year.append(i.split("-")[0])  # Extract the year part
    return year

# Apply the function to extract the year and create a new 'Year' column
df['Year'] = Extract_year()

# Group the data by year and calculate the success rate (mean of 'Class')
yearly_success_rate = df.groupby('Year')['Class'].mean().reset_index()

# Plot the line chart
plt.figure(figsize=(10,6))
plt.plot(yearly_success_rate['Year'], yearly_success_rate['Class'], marker='o', linestyle='-', color='b')

# Customize the chart
plt.xlabel('Year', fontsize=15)
plt.ylabel('Success Rate', fontsize=15)
plt.title('Yearly Success Rate of SpaceX Launches', fontsize=18)

# Display the plot
plt.show()

```

you can observe that the sucess rate since 2013 kept increasing till 2020


## Features Engineering


By now, you should obtain some preliminary insights about how each important variable would affect the success rate, we will select the features that will be used in success prediction in the future module.


```python
features = df[['FlightNumber', 'PayloadMass', 'Orbit', 'LaunchSite', 'Flights', 'GridFins', 'Reused', 'Legs', 'LandingPad', 'Block', 'ReusedCount', 'Serial']]
features.head()
```

### TASK  7: Create dummy variables to categorical columns


Use the function <code>get_dummies</code> and <code>features</code> dataframe to apply OneHotEncoder to the column <code>Orbits</code>, <code>LaunchSite</code>, <code>LandingPad</code>, and <code>Serial</code>. Assign the value to the variable <code>features_one_hot</code>, display the results using the method head. Your result dataframe must include all features including the encoded ones.


```python
# HINT: Use get_dummies() function on the categorical columns
# Import pandas if not already done
import pandas as pd

# Apply get_dummies to the categorical columns
features_one_hot = pd.get_dummies(features, columns=['Orbit', 'LaunchSite', 'LandingPad', 'Serial'])

# Display the first few rows of the dataframe
features_one_hot.head()
```
